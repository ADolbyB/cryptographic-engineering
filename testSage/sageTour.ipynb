#### Find Greatest Common Denominator: `gcd(x, y)`

```sage
a = 100
b = 25
d = gcd(a, b)
print(f'GCD of {a} and {b} = {d}\n')

e = 110
f = 22
g = gcd(e, f)
print(f'GCD of {e} and {f} = {g}\n')

h = 110
j = 21
k = gcd(h, j)
print(f'GCD of {h} and {j} = {k}\n')
```

#### Some Other Basic Functions:

```sage
# Extended Euclidean algorithm parameter of two integers
c = xgcd(a,b)
print(f'The XGCD of {a} and {b} = {c}\n')

# Generate a list of bits for an integer
a = 16
c = a.bits()
print(f'The binary bits of {a} (Little Endian) = {c}\n')

# Generate a list of digits of base b (b=2 is the .bits function) for an integer
# and extend to size n (fill 0s)
a = 32
b = 2
n = 0
c = a.digits(base=b, padto=n)
print(f'{a} in decimal = {c} as a list in base {b}\n')

# Convert list back to Integer (Reverse of previous function)
a = Integer(c, base=b)
print(f'Convert {c} back to the integer value = {a}\n')

# Get number of bits of an Integer
c = a.nbits()
print(f'Number of binary bits needed to represent {a} = {c}\n')

# Get number of digits of an Integer with base 10
a = 9999
c = a.ndigits()
print(f'Number of decimal digits needed to represent {a} = {c}\n')

# Create Field F_p
p = 101
K = GF(p)
print(K, '\n')

# Convert to hex
print(f'{a} decimal converted to hexadecimal = {hex(a)}')
print(f'{a} decimal converted to hexadecimal (w/o \'0x\' prefix) = {a.hex()}\n')

# Produce some random integer
c = ZZ.random_element()
print(f'The Random Number = {c}\n')

# Sometimes values will be printed as fraction or irrational value. 
# If you want decimal approximation of value, use this function
c = numerical_approx(a)
print(f'numerical_approx({a}) = {c}')
```
